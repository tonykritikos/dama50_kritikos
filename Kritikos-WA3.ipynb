```python
J = diagonal_matrix(RDF, [RDF.random_element(1,5) for _ in range(5)])
B = random_matrix(RDF, 5)
A = B.transpose() * J * B
A[1,1] = 7
A[2,2] = 9
A.is_positive_definite()
```

```python
A.n(digits=4)
```

```python
# 7.b
```

```python
C = A.cholesky()
C.n(digits=4)
```

```python
(C * C.transpose()).n(digits=4)
```

```python
A.n(digits=4) == (C * C.transpose()).n(digits=4)
```
